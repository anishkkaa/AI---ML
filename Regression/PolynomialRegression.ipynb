```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
df= pd.DataFrame({
    'X' : list(range(-100, 100, 5)),# Dummy line to avoid syntax error
})
df.head()
```

```python
plt.plot(df['X'])
```

```python
df['X2'] = df['X'] ** 2
df['X3'] = df['X'] ** 3
df['X4'] = df['X'] ** 4

df.head()
```

```python
fig, (ax1, ax2 , ax3, ax4) = plt.subplots(1, 4, figsize=(20,5))
ax1.plot(df['X'], color='red', label='degree1',marker='o')
ax2.plot(df['X2'], color='blue', label='degree2',marker='o') 
ax3.plot(df['X3'], color='green', label='degree3',marker='o') 
ax4.plot(df['X4'], color='black', label='degree4',marker='o')  
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.show()
```
